```python
from ensure import ensure_annotations
@ensure_annotations
def process_payment(amount: float, item_id: int) -> str:
    return f"Processed payment of ${amount} for item #{item_id}."

# This works perfectly:
print(process_payment(19.99, 1042))

# This will trigger a runtime EnsureError because a string is passed instead of an integer:
print(process_payment(19.99, "one-thousand-forty-two"))
```
